<b style="background-color:#B8FFAA;">Importing packages:</b>

```python
from math import *
import numpy as np
import pandas as pd
from pathlib import Path
import datetime
from matplotlib import cm 
import matplotlib.pyplot as plt
```

<b style="background-color:#B8FFAA;">Let's have a look at how the data looks like:</b>

```python
input_path = Path('deaths.csv')
data = pd.read_csv(input_path, encoding='utf-8')

data.head()
```

```python
input_path_age = Path('age.csv')

age = pd.read_csv(input_path_age, encoding='utf-8')

age.head()
```

<b style="background-color:#B8FFAA;">How many entries are there and what are the variables?</b>

```python
data.count()
```

<b style="background-color:#B8FFAA;">First of all, we can go ahead and delete the 'Code' column as the country name suffices. Speaking of which, we should change the 'Entitiy' column into 'Country' instead:</b>

```python
data = data.drop(['Code'],axis=1)
data = data.drop(['Number of executions (Amnesty International)'],axis=1)
data = data.rename(columns = {'Entity':'Country'})
data.head()
```

```python
age = age.drop('Code',axis=1)
age = age.rename(columns ={'Entity':'Country'})
```

<b style="background-color:#B8FFAA;">Now, for this data, the ages and sex are all inclusive in each row, so we only need the cause of death as the name for cleanliness:</b>

```python
data.columns = data.columns.str.replace('Deaths - ','')
data.columns = data.columns.str.replace(' - Sex: Both','')
data.columns = data.columns.str.replace(' - Age: All Ages','')
data.columns = data.columns.str.replace("\(Number\)",'',regex=True)
data.columns = data.columns.str.lstrip()
data.head()
```

```python
age.columns = age.columns.str.replace('Deaths - All causes - Sex: Both - Age:','')
age.columns = age.columns.str.replace(' \(Number\)','')
age.head()
```

<b style="background-color:#B8FFAA;">Clean up any null values in the dataset: </b>

```python
data = data.fillna(0)
```

```python
age = age.fillna(0)
```

<b style="background-color:#B8FFAA;">Let's correct the data types:</b>

```python
data.dtypes
```

```python
data.iloc[:,2:].astype('int64').head()
```

<h1>4. Exploratory data analysis</h1>

<h2>4.1 Leading Causes of Death (Overall)</h2>

<b style="background-color:#B8FFAA;">Let's start out by having a general look of the leading cause of deaths.</b>

```python
general = data.iloc[:,2:].sum().astype('int64').sort_values()
general
```

<b style="background-color:#B8FFAA;">To be more correct to the context of the data, let's normalize the values:</b>

```python
general = general/general.sum()
general
```

<b style="background-color:#B8FFAA;">Let's see the data in a bar graph:</b>

```python
general.plot.barh(figsize=(10,10))
plt.title('Leading cause of Death (Overall - World)')
plt.ylabel('Causes')
plt.xlabel('Deaths (percentage)')
plt.show()
```

<b style="background-color:#B8FFAA;">To get a better look at the percentages, let's switch to a pie chart:</b> 

```python
general.plot.pie(figsize=(10,10),
                 shadow=True,
                 explode=general/2,
                 colors=cm.Spectral_r(general),
                 autopct='%1.0f%%'
                 )
plt.show()
```

<b style="background-color:#B8FFAA;">The lower causes are cluttering, so let's add them up into 'others' for now:</b> 

```python
general.loc['others'] = general[general<0.01].sum()
```

```python
general = general.drop(general[general<0.01].index)
```

```python
general.plot.pie(figsize=(10,10),
                 shadow=True,
                 explode=general/2,
                 colors=cm.Spectral_r(general),
                 autopct='%1.0f%%'
                 )
plt.show()
```

<b style="background-color:#B8FFAA;">It appears that cardiovascular diseases and neoplasms(cancer) are the leading causes of death in the world, followed by repiratory and digestive diseases, with the leading two taking up over 47% of deaths. </b> 

<b style="background-color:#B8FFAA;">Let's have a look at what this looks like in Australia: </b> 

```python
australia = data[data['Country'].str.contains('Australia')].iloc[:,2:].sum().astype('int64').sort_values()
#normailzed:
australia = australia/australia.sum()
#adding 'others' column:
australia.loc['others'] = australia[australia<0.01].sum()
#deleting causes less than 1%:
australia = australia.drop(australia[australia<0.01].index)
```

```python
australia.plot.pie(figsize=(10,10),
                 shadow=True,
                 explode=australia/2,
                 colors=cm.Spectral_r(australia),
                 autopct='%1.0f%%'
                 )
plt.show()
```

<b style="background-color:#B8FFAA;">Upon quick inspection, Australia has huge jump in deaths caused by neoplasms(cancer) and a similar rate toward cardiovascular diseases. What does this indicate? Hence, the more preventable diseases are reduced and the proportion of the latter are almost like a constant, prevalent in all countries.

But why cancer? From a quick google search, the country with the highest rate of cancer(especially skin cancer) is none other than Australia. Another reason is due to the influence of population growth - as the proportion of older people grow, the number of cancer cases increase linearly. <b style="background-color:#B8FFAA;">
Back to the general view: let's calculate which way you are most likely to die from:
</b>

<b style="background-color:#B8FFAA;">
First, I will calculate all combinations of age group and causes, then calculate their probabilities:
    
</b>

```python
age_groups = age.drop(['Country','Year'],axis=1)
data_causes = data.drop(['Country','Year'],axis=1)
```

```python
age_list = list(age_groups)
cause_list = list(data_causes)

possibilities = []
for i in cause_list:
    for j in age_list:
        possibilities.append(i+" : "+j)
        
possibilities
```

<b style="background-color:#B8FFAA;">
Now, the probabilities:
</b>

```python
age_group_probability = age_groups.sum()/age_groups.sum().sum()
age_group_probability
```

```python
data_causes_probability = data_causes.sum()/data_causes.sum().sum()
data_causes_probability
```

```python
age_group_probability_list = list(age_group_probability)
age_group_probability_list
```

```python
data_causes_probability_list = list(data_causes_probability)
data_causes_probability_list
```

```python
possibility_probability = []

for i in data_causes_probability_list:
    for j in age_group_probability_list:
        possibility_probability.append(i*j)
```

```python
possibility_probability
```

```python
combination = {possibilities[i] : possibility_probability[i] for i in range(len(possibilities))}
combination
```

<b style="background-color:#B8FFAA;">
And Finally, The dataframe:
</b>

```python
combination_df = pd.DataFrame.from_dict(combination,orient='index')
combination_df.columns=['probability']
combination_df.sort_values(by='probability',ascending=False).head(10)
```

<b style="background-color:#B8FFAA;">
Now we have all possibilites with their probabilities and we can observe that the most likely death a person from a sample from the world will have is through dardiovascular diseases at 70 years +.
</b>
